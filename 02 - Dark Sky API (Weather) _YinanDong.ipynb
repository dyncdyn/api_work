## 5) Go through the daily forecasts, printing out the next week's worth of predictions.

I'd like to know the **high temperature** for each day, and whether it's **hot, warm, or cold** (based on what temperatures you think are hot, warm or cold).

- *Tip: Is it returning F or C? You can change them by adding `?units=us` or `?units=si` onto the end of the URL.*
- *Tip: You'll need to use an `if` statement to say whether it is hot, warm or cold.*

```python
Daily = Home['daily']['data']
for day in Daily:
    hightemp = day['temperatureHigh']
    if hightemp > 80:
        print(f'The high temperature is {hightemp}, and it is hot.')
    elif hightemp < 50:
        print(f'The high temperature is {hightemp}, and it is cold.')
    else:
        print(f'The high temperature is {hightemp}, and it is warm.')
```

# 6) What will be the hottest day in the next week? What is the high temperature on that day?

It's okay if your result looks like `1453363200`. If you want to be extra-fancy, though, format it as a **real date** - the weird number is called a **UNIX timestamp**, which might help in googling.

- *Tip: It's okay to skip this if it seems too difficult!*

```python
Date = []
HighTemp = []

for day in Daily:
    date = day['time']
    Date.append(date)
    hightemp = day['temperatureHigh']
    HighTemp.append(hightemp)

NextWeek = list(zip(HighTemp,Date))
Hottest = max(NextWeek)

Hottest
```

## 7) What's the weather looking like for the next 24+ hours in Miami, Florida?

I'd like to know the temperature for every hour, and if it's going to have cloud cover of more than 50% say "{temperature} and cloudy" instead of just the temperature. 


- *Tip: Be sure to remember how latitude and longitude are done with positive and negative numbers*
- *Tip: How do you represent "50%" for cloud cover?*

```python
response = requests.get('https://api.darksky.net/forecast/80f9cf4395301f0ba1ebadb4f00716aa/25.7617,-80.1918')
Miami = response.json()
Miami
```

```python
Miami24 = Miami['hourly']['data']
Miami24
```

```python
for hour in Miami24:
    temp = hour['temperature']
    if hour['cloudCover'] > 0.5:
        print(temp,"and cloudy")
    else:
        print(temp)
        
```

# 8) For the next 24-ish hours in Miami, what percent of the time is the temperature above 85 degrees?

If they give you 48 hours instead of 24, that's okay.

- *Tip: You might want to read up on [looping patterns](http://jonathansoma.com/lede/foundations-2017/classes/data%20structures/looping-patterns/)*

```python
Above85 = []
MiamiTemp = []
for hour in Miami24:
    temp = hour['temperature']
    MiamiTemp.append(temp)
    if hour['temperature'] > 85:
        Above85.append(temp)
Percentage = f'{round(len(Above85)/len(MiamiTemp)*100,2)}%'
Percentage
```
